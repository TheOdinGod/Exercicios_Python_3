```python
## importando bibliotecas
import pandas as pd
import seaborn as srn
import statistics as sts
import matplotlib.pyplot as plt
```

```python
#baixando a base de dados
df = pd.read_csv("/content/enhanced_anxiety_dataset.csv")
df.shape
```

```python
#visualizar as primeiras linhas
df.head()
```

```python
#visualizar as ultimas linhas
df.tail()
```

```python
#agrupando dados
df_occupation = df.groupby(["Occupation"]).size()
df_occupation
```

```python
#grafico de barras
df_occupation.plot.bar()
```

```python
#comparar as profissões
group_occupation_anxiety = df.groupby(["Occupation"])["Anxiety Level (1-10)"].mean()
group_occupation_anxiety.plot.bar()

```

```python
#profissões x consumo de cafeína
group_occupation_anxiety = df.groupby(["Occupation"])["Anxiety Level (1-10)"].sum()
group_occupation_anxiety.plot.bar()
```

```python
#grafico de barras
df_occupation.plot.pie()
```

```python
#grafico por genero
group_gender = df.groupby(['Gender']).size()
group_gender.plot.pie()
```

```python
## Histogramas
srn.histplot(df["Sleep Hours"],
             kde = True,
             bins = 15,
             color = "purple").set_title("Horas de Sono")
```

```python
## só a linha de densidade
srn.kdeplot(df["Sleep Hours"])
```

```python
# histograma da atividade fisica (Physical Activity (hrs/week))
srn.histplot(df["Physical Activity (hrs/week)"],
             kde = True,
             bins = 10,
             color = "purple").set_title("Horas de Atividade Física")

```

```python
# histograma comparando dois atributos
# engenheiros x horas sono
srn.histplot(
    df.loc[df["Occupation"] == "Engineer"]["Sleep Hours"]
).set_title("Horas de Sono dos Engenheiros")
```

```python
# Horas de sono dos artistas
srn.histplot(
    df.loc[df["Occupation"] == "Artist"]["Sleep Hours"]
).set_title("Horas de Sono dos Artistas")
```
