```python
import numpy as np

# 1. Converter array para float
array1 = np.array([3, 5, 9, 16, 25], dtype=float)
print(array1)

# 2. Converter array para int
array2 = np.array([1.2, -3.6, 8.99999, 16.1], dtype=int)
print(array2)

# 3. Criar matriz 3x9 de zeros
matriz_zeros = np.zeros((3, 9), dtype=int)
print(matriz_zeros)

# 4. Criar matriz 6x7 de uns
matriz_uns = np.ones((6, 7), dtype=int)
print(matriz_uns)

# 5. Criar matriz identidade 4x4
matriz_identidade = np.identity(4, dtype=int)
print(matriz_identidade)

# 6. Criar array de números aleatórios entre 0 e 1
array_aleatorio = np.random.rand(5)
print(array_aleatorio)

# 7. Criar matriz 3x6 de números aleatórios inteiros de 0 a 5
matriz_aleatoria_inteiros = np.random.randint(0, 6, size=(3, 6))
print(matriz_aleatoria_inteiros)

# Matriz para os exercícios 8 a 11
matriz_exercicios = np.array([[4, 6, 8], [-4, 3, 2], [22, -15, 12]])

# 8. Valor máximo da matriz
valor_maximo = np.max(matriz_exercicios)
print(valor_maximo)

# 9. Valor mínimo da matriz
valor_minimo = np.min(matriz_exercicios)
print(valor_minimo)

# 10. Soma dos elementos da matriz
soma_elementos = np.sum(matriz_exercicios)
print(soma_elementos)

# 11. Média dos elementos da matriz
media_elementos = np.mean(matriz_exercicios)
print(media_elementos)

# Array para os exercícios 12 a 16
array_exercicios = np.array([13, -2, 6, 8, 23, 15, 3, 90, 4, 10, 56, 12])

# 12. 4º elemento do array
quarto_elemento = array_exercicios[3]
print(quarto_elemento)

# 13. 3 últimos elementos do array
ultimos_tres = array_exercicios[-3:]
print(ultimos_tres)

# 14. Elementos a partir da posição 2
apartir_posicao_2 = array_exercicios[2:]
print(apartir_posicao_2)

# 15. Array entre o 2º e o 6º elementos
entre_2_e_6 = array_exercicios[1:6]
print(entre_2_e_6)

# 16. Array [23 15 3 90]
array_especifico = array_exercicios[4:8]
print(array_especifico)

# Matriz para os exercícios 17 a 20
matriz_exercicios_2 = np.array([[1, 3, 16, -25, 14], [2, 0, 10, 5, 17], [19, 4, 8, 15, 13], [7, 12, -5, 14, 0]])

# 17. 3ª linha da matriz
terceira_linha = matriz_exercicios_2[2, :]
print(terceira_linha)

# 18. 4ª coluna da matriz
quarta_coluna = matriz_exercicios_2[:, 3]
print(quarta_coluna)

# 19. Elemento de valor 10
elemento_10 = matriz_exercicios_2[1, 2]
print(elemento_10)

# 20. Matriz [[0 10 5] [4 8 15]]
matriz_especifica = matriz_exercicios_2[1:3, 2:5]
print(matriz_especifica)

# Matriz para o exercício 21
matriz_exercicio_21 = np.array([[2, -1, 15, 24], [-3, 18, 4, 25], [9, 10, -13, 7]])

# 21. Criar nova matriz com 1 para ímpares e 2 para pares
matriz_pares_impares = np.where(matriz_exercicio_21 % 2 == 0, 2, 1)
print(matriz_pares_impares)
```
